```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cufflinks as cf
from IPython.display import display,HTML
```

```python
#Leo mi archivo.csv
df = pd.read_csv("CountryTable.csv")
df
```

### ***REEMPLAZAR VALORES NULOS***

```python
#Obtenemos la cuenta de los elementos nulos del dataframe por COLUMNA
Cuenta_Nulos= df.isnull().sum()
Cuenta_Nulos
```

```python
## “independence_year”: Rellenar con el valor “1900”
#Sustituir valores nulos por un valor numérico en  concreto 
df["independence_year"]=df["independence_year"].fillna(1900)
df
```

```python
# Comprobamos valores nulos
Cuenta_Nulos= df.isnull().sum()
Cuenta_Nulos
```

```python
## “life_expectancy”: Rellenar con el valor “Media” de la columna
df["life_expectancy"]=df["life_expectancy"].fillna(round(df["life_expectancy"].mean(),1))
Cuenta_Nulos= df.isnull().sum()
Cuenta_Nulos
```

```python
## “gnp_old ”: Rellenar con el valor “Mediana” de la columna
#Sustituir valores nulos con mediana
df["gnp_old"]=df["gnp_old"].fillna(round(df["gnp_old"].median(),1))
Cuenta_Nulos= df.isnull().sum()
Cuenta_Nulos
```

```python
## “local_name”: Rellenar con el string “Inexistente”
df["local_name"]=df["local_name"].fillna("Inexistente")
Cuenta_Nulos= df.isnull().sum()
Cuenta_Nulos
```

```python
## “head_of_state”: Rellenar con el string “Desconocido”
df["head_of_state"]=df["head_of_state"].fillna("Desconocido")
Cuenta_Nulos= df.isnull().sum()
Cuenta_Nulos
```

```python
## “capital ”: Rellenar con el valor “0”
df["capital"]=df["capital"].fillna(0)
Cuenta_Nulos= df.isnull().sum()
Cuenta_Nulos
```

```python
## “code2”: Rellenar con el string “NA”
df["code2"]=df["code2"].fillna("NA")
Cuenta_Nulos= df.isnull().sum()
Cuenta_Nulos
```

### ***OUTLIERS***

Aplicar método de distribución normal para identificar Outliers, tomando como limites 2 desviaciones estándar, exclusivamente para las columnas “gnp old” y “population”, obtener los diagramas de caja de cada una respectivamente.

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
df_prueba = df[["gnp_old","population"]]
df_prueba
```

```python
#diagrama de caja
fig = plt.figure(figsize =(15, 8))
df_prueba.plot(kind='box')
plt.title("Valores atípicos")
plt.show()
```

```python
## Diagrama de caja "gnp_old"
fig = plt.figure(figsize =(5, 3))
plt.boxplot(df["gnp_old"]) 
plt.title("Outliers de GNP_OLD")
plt.show() #dibujamos el diagrama
```

```python
y=df["gnp_old"]
Limite_Superior= y.mean() + 2*y.std()
Limite_Inferior= y.mean() - 2*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers
outliers= df[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos limpios
data_clean= df[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data_clean
```

```python
## Diagrama de caja "gnp_old" terminado
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean["gnp_old"]) 
plt.title("Outliers de GNP_OLD")
plt.show() #dibujamos el diagrama
```
