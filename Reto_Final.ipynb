```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cufflinks as cf
from IPython.display import display,HTML
```

```python
#Leo mi archivo.csv
df = pd.read_csv("CountryTable.csv")
df
```

### ***REEMPLAZAR VALORES NULOS***

```python
#Obtenemos la cuenta de los elementos nulos del dataframe por COLUMNA
Cuenta_Nulos= df.isnull().sum()
Cuenta_Nulos
```

```python
## “independence_year”: Rellenar con el valor “1900”
#Sustituir valores nulos por un valor numérico en  concreto 
df["independence_year"]=df["independence_year"].fillna(1900)
df
```

```python
# Comprobamos valores nulos
Cuenta_Nulos= df.isnull().sum()
Cuenta_Nulos
```

```python
## “life_expectancy”: Rellenar con el valor “Media” de la columna
df["life_expectancy"]=df["life_expectancy"].fillna(round(df["life_expectancy"].mean(),1))
Cuenta_Nulos= df.isnull().sum()
Cuenta_Nulos
```

```python
## “gnp_old ”: Rellenar con el valor “Mediana” de la columna
#Sustituir valores nulos con mediana
df["gnp_old"]=df["gnp_old"].fillna(round(df["gnp_old"].median(),1))
Cuenta_Nulos= df.isnull().sum()
Cuenta_Nulos
```

```python
## “local_name”: Rellenar con el string “Inexistente”
df["local_name"]=df["local_name"].fillna("Inexistente")
Cuenta_Nulos= df.isnull().sum()
Cuenta_Nulos
```

```python
## “head_of_state”: Rellenar con el string “Desconocido”
df["head_of_state"]=df["head_of_state"].fillna("Desconocido")
Cuenta_Nulos= df.isnull().sum()
Cuenta_Nulos
```

```python
## “capital ”: Rellenar con el valor “0”
df["capital"]=df["capital"].fillna(0)
Cuenta_Nulos= df.isnull().sum()
Cuenta_Nulos
```

```python
## “code2”: Rellenar con el string “NA”
df["code2"]=df["code2"].fillna("NA")
Cuenta_Nulos= df.isnull().sum()
Cuenta_Nulos
```

Aplicar método de rango intercuartílico para identificar Outliers, tomando como limites 1.3 iqr, exclusivamente para las columnas “gnp old”, “surface área”, “life expectancy” y “population”, obtener los diagramas de caja de cada una respectivamente

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y3=df["gnp_old"]

percentile25=y3.quantile(0.25) #Q1
percentile75=y3.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.3*iqr
Limite_Inferior_iqr= percentile25 - 1.3*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Ouliers
outliers_iqr= df[(y3>Limite_Superior_iqr)|(y3<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios
data_clean_iqr= df[(y3<=Limite_Superior_iqr)&(y3>=Limite_Inferior_iqr)]
data_clean_iqr
```

```python
#diagrama de caja de gnp_old
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean_iqr["gnp_old"]) 
plt.title("Outliers de GNP_OLD")
plt.show() #dibujamos el diagrama
```

### ***Transformaciones y representación de datos***

Aplica las siguientes transformaciones al DataFrame original y obtener en cada una las siguientes representaciones de datos: Grafico Lineal (Lineplot), Grafico de barras (Barplot), Grafico de cajas, (Boxplot) Histograma (Histplot) y Mapa de dispersión (Scatterplot) 

```python
cf.set_config_file(sharing='public',theme='pearl',offline=True) # write cf.getThemes() to find all themes available
```

```python
data = df.copy()
```

```python
#primera reestructuración
data1=data.pivot(index='code', columns='continent',  values='population')
data1
```

```python
#Lineplot o grafico lineal
data1.iplot(kind='line', xTitle='continente', yTitle='población', title='población por continentes')
```

```python
#Barplot
data1.iplot(kind='bar', xTitle='continente', yTitle='población', title='población por continentes')
```

```python
#Grafico de cajas, (Boxplot)
data1.iplot(kind='box', xTitle='continente', yTitle='población', title='Población por continentes')
```

```python
# Histograma (Histplot)
data1.iplot(kind='hist', xTitle='poblacion', title='distribución de población')
```

```python
## Mapa de dispersión (Scatterplot)
data1.iplot(kind='scatter', mode='markers',xTitle='code of continent', 
                               title='Distibución de población por continente')
```
